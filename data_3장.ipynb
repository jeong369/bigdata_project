## 시카고 샌드위치 가게이름, 가게 메인메뉴, 가게 소개 페이지 정리

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen, urljoin

url = "https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/"
page = urlopen(url)

soup = BeautifulSoup(page, 'html.parser')

soup
```

```python
print(len(soup.body.find_all('div', 'sammy'))) # 50개 맛집 맞다!
print(soup.body.find_all('div', 'sammy'))
```

```python
print(soup.body.find_all('div', 'sammy')[0])
```

```python
tmp_one = soup.find_all('div', 'sammy')[0]
type(tmp_one)

# type이 bs4.element. Tag 는 다시 find/find_all로 찾을 수 있다.
```

```python
print(tmp_one.find_all('a')[0]['href'])
tmp_one.find_all('b')[0].get_text()
```

```python
# 정규식 (Regular Express)

import re

tmp_string = tmp_one.find(class_='sammyListing').get_text()
tmp_string

print('가게이름 : ', re.split(('\n|\r\n'), tmp_string)[1])
print('메인메뉴 : ', re.split(('\n|\r\n'), tmp_string)[0])
```

- url은 상대경로와 절대경로가 있다. ex)방금 위는 상대 경로
- urljoin : 절대경로는 그대로 두고, 상대경로를 절대경로로 바꾸어준다.

```python
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy')

for item in list_soup :
    rank.append(item.find(class_="sammyRank").get_text())
    tmp_string = item.find(class_="sammyListing").get_text()
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    url_add.append(urljoin(url, item.find('a')['href']))
```

```python
url_add[:5]
```

```python
# pandas를 이용하여 리스트에 저장한다.

import pandas as pd

data = {'Rank':rank, 'Cafe':cafe_name, 'Menu':main_menu, 'URL': url_add}
df = pd.DataFrame(data)
df.head()
```

```python
# 저장
df.to_csv('../data/03. best_sandwiches_list_chicago.csv', sep=','
         , encoding='UTF-8')
```

```python
lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

## 네이버 영화 평점 기준 영화의 평점 변화 확인하기

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from urllib.request import urlopen

url_base = "http://movie.naver.com/"
url_sub = "movie/sdb/rank/rmovie.nhn?sel=cur&date=20190529"

page = urlopen(url_base + url_sub)
soup = BeautifulSoup(page, "html.parser")

soup
```

```python
soup.find_all('div', 'tit5')
```

```python
soup.find_all('div','tit5')[0].a.string
```

```python
soup.find_all('td','point')[0].string
```

```python
date = pd.date_range('2019-5-1', periods=100, freq='D')
date
```

```python
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date):
    html = "http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, 'html.parser')
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
    
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie['point'] = movie['point'].astype(float)
movie.head()
```

```python
import numpy as np

movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
tmp = movie.query('name == ["어벤져스: 엔드게임"]')
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')
plt.grid()
plt.show()
```
